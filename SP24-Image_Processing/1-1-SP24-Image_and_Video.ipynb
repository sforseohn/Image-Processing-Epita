##### <font color="dodgerblue">**What's going on with the color?**</font> <font size="5">&#x1F631;</font>

This is the effect of [color mapping](https://en.wikipedia.org/wiki/Image_color_transfer).
Indeed, <font size='2'>📚</font> [matplotlib.pyplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html) is designed to display various types of 2D or 3D data (e.g. array, matrix, image, histogram, ...) and associates some colors with the values.
What we see on the displayed image are then fake mapped colors.

Those colors are not random but defined by a function mapping a specific color to each value of the data.
The main idea behind this is to give the opportunity to use the right mapping function to highlight the relevant informations stored in the data.

Here is a simple hand-made example of [color mapping](https://en.wikipedia.org/wiki/Image_color_transfer):

| Values | Colormap |
|---:|---:|
| 0 | red |
| 1 | orange |
| 2 | green |

Which means that all ``0`` values will be shown in ``red``, all ``1`` in ``orange`` and all ``2`` in ``green``.

We can also do a mapping by range:

| Range of values | Colormap |
|---:|---:|
| [0, 10] | blue |
| [11, 24] | orange |
| [25, 40] | red |

We can, for example, think that:
- the temperatures between ``0`` and ``10`` are cold enough to be represented by the ``blue`` color,
- the temperatures between ``11`` and ``24`` are warmer so represented by the ``orange``
- and those between ``25`` and ``40`` are hot enough to be associated with the ``red`` color.

Many predefined [colormaps](https://matplotlib.org/stable/users/explain/colors/colormaps.html) are available in <font size='2'>📚</font> [matplotlib](https://matplotlib.org/stable/api/matplotlib_configuration_api.html#matplotlib.colormaps).
They can directly be used by specifying the ``cmap`` parameter of the <font size='2'>⚙</font> [imshow](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html) funcion.

If ``cmap`` is not specified, the default one which is <font color='red'>'viridis'</font> is applied.

#### <font color="dodgerblue">1.1.6. Image encoding</font>

##### <font color="dodgerblue">**Why are the possible values of our pixel between ``0`` and ``255``?**</font>

In order to store the data contained by each pixel of an image, an encoding step is required. It defines how the values are stored in the matrix and which [image file format](https://en.wikipedia.org/wiki/Image_file_format) to use later (jpeg, gif, png,...) for a saving process.

In our case, we loaded the grayscale image ``gray1.png`` having a ``PNG`` or Portable Network Graphics file format which is widely used and well-known. 

The ``PNG`` format can support different types of encoding but we obviously have here an **8-bit** palette image that leads to **256** possible levels for each pixel.
Yes, from ``0`` to ``255`` both included.

See the table bellow for few examples giving the intuition on how many possible values a pixel can hold for each number of bits considered during the image encoding.

| Number of bits   |      Number of possible values      |
|----------:|-------------:|
| 2-bit | 2 |
| 8-bit  | 256 |
| 16-bit | 65,536   |
| 24-bit | 16,7 million   |
| 32-bit | 1,07 billion |

##### <font color="dodgerblue">**Why are all the values positive integer?**</font>

Let's see the loaded matrix ``gray_image`` again. <font size='5'>&#x1F440;</font> 

````python
gray_image
````

When we look at the matrix, we notice that the values are only [positive integers](https://en.wikipedia.org/wiki/Integer).

Indeed, it is confirmed by the ``dtype`` shown at the bottom right of the printed matrix.
This is the [data type](https://en.wikipedia.org/wiki/Data_type) used during the encoding step.

``uint8`` means here that the pixel values are encoded using ``8 bits``, they are ``integers`` and ``unsigned``, i.e. only positive sign.


We also can get this ``data type`` information by doing as bellow.

````python
gray_image.dtype
````

Many ``data types`` as ``uint16``, ``uint32``, ``uint64``, ``float64`` or even ``bool`` (for binary images) exists but we will only consider ``uint8`` or ``ubyte`` which will give us matrices having possible values between ``[0, 255]`` which is the range adapted to the human natural vision. <font size='5'>	&#x1F441; 	&#x1F441;</font>

````python
height, width = gray_image.shape
print(f"height: {height}, width: {width}")
````
